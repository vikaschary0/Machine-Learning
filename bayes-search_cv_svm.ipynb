```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer
```

```python
df = pd.read_csv('/kaggle/input/iris-dataset/iris.csv')
```

```python
X = df.drop('species', axis=1)  
y = df['species']
```

```python
param_space = {
    'C': Real(1e-6, 1e+6, prior='log-uniform'),
    'gamma': Real(1e-6, 1e+1, prior='log-uniform'),
}
```

```python
svm = SVC(kernel='linear')  # You can choose other kernels like 'linear', 'poly', etc.

bayes_search = BayesSearchCV(
    estimator=svm,
    search_spaces=param_space,
    n_iter=50,  # Number of optimization iterations
    cv=5,  # Number of cross-validation folds
    n_jobs=-1,  # Use all available CPU cores
    verbose=1,  # Increase verbosity for more information
    refit=True,  
    random_state=42,  
)
```

```python
bayes_search.fit(X, y)
```

```python
best_params = bayes_search.best_params_
best_score = bayes_search.best_score_
best_estimator = bayes_search.best_estimator_
print("best parameters:",best_params)
print("best score:",best_score)
print("best estimator:",best_estimator)
```
